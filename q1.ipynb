```python
# calculate the deriviative 
def grad_f(x):
    return 4+2*1*x
```

## 4

```python
# finding the minimum with Gradiant decent Algorithem
def grad_update(grad, x, learning_rate_parm):
    x_updated = x-learning_rate_parm*grad
    return x_updated
```

## 5

```python

# calculate the gradiant decent algorithem
def gd_alg(x_0,epsilon,learning_rate_parm):
    x_current=x_0

    x_new=grad_update(grad_f(x_current),x_current,learning_rate_parm)
    updates_x_list=list()
    updates_x_list.append(x_current)

    while abs(x_new-x_current)>epsilon:
        x_current=x_new
        x_new=grad_update(grad_f(x_current),x_current,learning_rate_parm)
        updates_x_list.append(x_new)
    return updates_x_list 
  
```

```python

x_list=gd_alg(15 ,0.000001, 0.01)
print("The X value that the algorithem converged to is : "+str(x_list[-1]))    
```

The x value we got is very close to the actual critical point (-2) but it's not the same,
 it is because we make approximation and have always $\epsilon>0$, note that if we wiil minimize $\epsilon$ 
such that it approximates to 0 then the algorithem returns closer answer to the actual critical point (-2)


## 6

```python
import math
# finding the optimal Hyperparameters for convergence in the smallest number of iterations
t_min=math.inf
x_0_list=np.arange(20,10,-1)
epsilon_list=np.arange(0.01,0.001,-0.001)
learning_rate_parm_list=np.arange(0.01,0.001,-0.001)
for epsilon in epsilon_list:
    for learning_rate_parm in learning_rate_parm_list:
        for x_0 in x_0_list:
            reuslt=gd_alg(x_0,epsilon,learning_rate_parm)
            x_convergence=reuslt[-1]
            T=len(reuslt)
            if T<t_min:
                min_result=reuslt
                t_min=T
                x_0_min=x_0
                epsilon_min=epsilon
                learning_rate_parm_min=learning_rate_parm
print("The Best hyper parameters for the fastest convergence are: x_0 = "+str(x_0_min)+", epsilon = "+str(epsilon_min)+", learning_rate_parm = "+str(learning_rate_parm_min))
print("The X value that the algorithem converged to is : "+str(x_convergence), "The number of iterations is: "+str(t_min))
```
